## **TF-IDF를 조금 다르게 정의하는 것도 가능하다.**



TF- IDF는 단어의 빈도와 역 문서 빈도를 사용하여 DTM 내의 각 단어들마다 중요한 정도를 가중치로 주는 방법이다

TF-IDF는 단어의 빈도와 역 문서 빈도를 사용하여 DTM 내의 각 단어들마다 중요한 정도를 가중치로 주는 방법이다.

우선 DTM을 만든 후, TF-IDF의 가중치를 부여한다.

TF-IDF는 주로 문서의 유사도를 구하는 작업, 검색 시스템에서 검색 결과의 중요도를 정하는 작업, 문서 내에서 특정 단어의 중요도를 구하는 작업 등에 쓰일 수 있다.

문서를 d, 단어를 t, 문서의 총 개수를 n이라 표현할 때, TF,DF,IDF는 다음과 같이 정의할 수 있다.

### 1) tf(d,t): 특정 문서 d에서의 특정 단어 t의 등장 횟수

TF는 앞에서 구한 적이 있다. TF는 앞에서 배운 DTM의 예제에서 각 단어들이 가진 값이다.
DTM이 각 문서에서의 각 단어의 등장 빈도를 나타내는 값이기 때문이다.

### 2) df(t) : 특정 단어 t가 등장한 문서의 수

특정 단어가 각 문서 또는 문서들에서 몇 번 등장했는지는 관심사항이 아니며, 특정 단어 t가 등장한 문서의 수에만 관심을 가진다.

앞에서 배운 DTM을 다시 보면 바나나는 문서 2와 문서 3에서 등장했다.

\



이 경우, 바나나의 df는 2이다.

문서 3에서 바나나가 두 번 등장했지만, 중요하지 않다.

심지어 한 문서에서 100번 등장하던 1000번 등장하던, 등장한 문서의 수에만 관심을 갖는다

## 3) idf(d,t) : df(t)에 반비례하는 수




IDF는 DF의 역수의 꼴은 맞으나, log를 취하는 이유는 IDF를 DF의 역수로 사용한다면, 총 문서의 수 n이 커질 수록, IDF의 값은 기하급수적으로 커진다.\
그렇기에 log를 사용한다.

직관적으로는 불용어 등과 같이 자주 쓰이는 단어들은 비교적 자주 쓰이지 않는 단어들보다 최소 수십 배 자주 등장한다.\
비교적 자주 쓰이지 않는 단어들조차 희귀 단어들과 비교하면 수백 배는 자주 등장하게 된다. 이러한 이유로 log를 씌워주지 않으면, 희귀 단ㅇ처들에 엄청난 가중치가 부여될 수 있다.

log를 씌우면 이러한 격차를 줄이는 효과가 존재한다. 또한 log 안의 식에서 분모에 1을 더해주는 이유는 특정 단어가 전체 문서에서 등장하지 않을 경우, 분모가 0이 되는 것을 방지하기 우ㅟ함.


TF-IDF는 모든 문서에서 자주 등장하는 단더는 중요도가 낮다고 판단하며, 특정 문서에서만 자주 등장하는 단어는 중요도가 높다고 판단한다.

TF-IDF 값이 낮으면, 중요도가 낮은 것이며, TF-IDF 값이 크면 중요도가 큰 것이다.

the, a 와 같은 불용어의 경우에는 모든 문서에 자주 등장하기 마련이기에 불용어의 TF-IDF의 값은 다른 단어의 TF-IDF에 비해 낮아지게 된다.

IDF를 구해야한다. 이때 사용하는 log는 TF-IDF를 사용하는 사용자가 임의로 설정이 가능하다. 

여기서 log는 기존의 값에 곱하여 값의 크기를 조절하는 상수의 역할을 수행하게 된다.\

각 단어에 대해서 IDF의 값을 비교해보면 문서 1개에만 등장한 단어와 문서 2개에만 등장한 단어는 값의 차이를 보인다.

IDF는 여러 문서에서 등장한 단어의 가중치를 낮추는 역할을 수행하게 된다.

TF = DTM에서의 각 단어의 값과 같으므로, 앞에서 사용한 DTM에서 단어 별로 위의 IDF 값을 곱해주면 TF-IDF의 값을 얻는다.

## python을 이용하여 TF-IDF 구현하기

위의 계산과정을 python을 이용하여 구현해볼 것. 앞의 설명에서 사용한 4개의 문서를 docs에 저장한다

```python
import pandas as pd
from math import log

docs = [
    '먹고 싶은 사과',
    '먹고 싶은 바나나',
    '길고 노란 바나나 바나나',
    '저는 과일이 좋아요'
]

vocab = list(set(w for doc in docs for w in doc.split()))
vocab.sort()
```

```python
# 총 문서의 수
N = len(docs)

def tf(t,d):
    return d.count(t)

def idf(t):
    df = 0
    for doc in docs:
        df += t in doc
    return log(N/(df+1))

def tfidf(t, d):
    return tf(t,d)* idf(t)
```

```python
result = []

# 각 문서에 대해서 아래 연산을 반복한다.

for i in range(N):
    result.append([])
    d = docs[i]
    for j in range(len(vocab)):
        t = vocab[j]
        result[-1].append(tf(t,d))
        
tf_ = pd.DataFrame(result, columns = vocab)
tf_
```

DTM이 출력된 것을 확인 할 수 있다. 각 단어에 대한 IDF의 값을 구해보자

```python
result = []
for j in range(len(vocab)):
    t = vocab[j]
    result.append(idf(t))
    
idf_ = pd.DataFrame(result, index=vocab, columns=["IDF"])
idf_
```

직접 구한 IDF 값과 동일하다. TF-IDF 행렬을 출력해보자

```python
result = []
for i in range(N):
    result.append([])
    d = docs[i]
    for j in range(len(vocab)):
        t = vocab[j]
        result[-1].append(tfidf(t,d))
        
tfidf_ = pd.DataFrame(result, columns = vocab)
tfidf_
```

위의 계산식에는 몇가지 문제점이 존재한다.

전체 문서의 수 n이 4인데, df(t)의 값이 3인 경우 어떤 일이 벌어질까

df(t)에 1이 더해지면서 log 항의 분자와 분모의 값이 같아진다. (log(N/(df+1)))

이 때, log 안의 값이 1이 되며 idf(d,t)의 값이 0이 된다. IDF의 값이 0이 되는 순간 가중치의 역할을 수행할 수 없다.

따라서 실제 TF-IDF를 구현하는 모델들은 조정된 식을 사용한다

## 3. scikit-learn을 이용한 DTM과 TF-IDF 

scikit-learn 을 이용할 때 CountVectorizer을 이용하면 DTM을 만들 수 있다.

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    'you know I want your love',
    'l like you',
    'what should I do'
]

vector = CountVectorizer()

# corpus로 부터 각 단어의 빈도수를 기록

print(vector.fit_transform(corpus).toarray())

# 각 단어와 mapping 된 index를 출력

print(vector.vocabulary_)
```

DTM에서 각 단어의 index가 어떻게 부여되었는지 확인하기 위해, index를 확인해 보았다.

첫번째 열의 경우, 0의 index를 가진 do 이다.

do는 세번째 문서에만 등장했기 때문에, 세번째 행에서만 1의 값을 가진다.

두번째 열의 경우에는 1의 index를 가진 know이다.

know는 첫번째 문서에만 등장했으므로 첫번째 행에서만 1의 값을 갖는다.


Scikit-learn은 TF-IDF를 자동 계산해주는 *TfidfVectorzier* 를 제공한다

scikit-learn의 TF-IDF는 위에서 배웠던 TF-IDF 기본 식에서 조정된 식을 사용한다.

IDF의 log항의 분자에 1을 더해주며, 로그항에 1을 더해주고, TF-IDF에 L2 정규화를 이용하여 값을 조정해준다.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    'you know I want your love',
    'l like you',
    'what should I do',
]

tfidfv = TfidfVectorizer().fit(corpus)
print(tfidfv.transform(corpus).toarray())
print(tfidfv.vocabulary_)
```
